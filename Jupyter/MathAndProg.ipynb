```python

import matplotlib.pyplot as plt
import numpy as np

axes = plt.gca()
axes.set_xlim(-7, 7)
axes.set_ylim(-5, 6)
axes.axhline(0, color='black', linewidth=0.5)  # 水平线
axes.axvline(0, color='black', linewidth=0.5)  #
dino_ve = [(6, 4), (3, 1), (1, 2), (-1, 5), (-2, 5), (-3, 4), (-4, 4)]
dino_vectors = [(6, 4), (3, 1), (1, 2), (-1, 5), (-2, 5), (-3, 4), (-4, 4), (-5, 3), (-5, 2), (-2, 2), (-5, 1), (-4, 0),
				(-2, 1), (-1, 0), (0, -3), (-1, -4), (1, -4), (2, -3), (1, -2), (3, -1), (5, 1), (6, 4)]
axes.plot([coord[0] for coord in dino_ve], [coord[1] for coord in dino_ve],'bo')
axes.plot([coord[0] for coord in dino_vectors], [coord[1] for coord in dino_vectors], 'bo-')
plt.show()
```

```python
import matplotlib.ticker
from matplotlib.pyplot import gca

axes = gca()
x_value = range(-9, 10)
y_value = [x ** 2 + x + 10 for x in x_value]
axes.set_xlim(min(x_value) - 1, max(x_value) + 1)
axes.set_ylim(0, max(y_value) + 1)
axes.axhline(0, color='black', linewidth=1)  # 水平线
axes.axvline(0, color='black', linewidth=1)  #
axes.xaxis.set_major_locator(matplotlib.ticker.MultipleLocator(1))
axes.plot(x_value, y_value, 'bo-')
plt.show()
```

```python
axes = gca()
axes.set_xlim(-10, 10)
axes.set_ylim(-10, 10)
axes.axhline(0, color='black', linewidth=0.5)  # 水平线
axes.axvline(0, color='black', linewidth=0.5)  #
dino_vectors2 = [np.add((-1.5, -2.5), v) for v in dino_vectors]
axes.plot([coord[0] for coord in dino_vectors], [coord[1] for coord in dino_vectors], 'bo-')
axes.plot([coord[0] for coord in dino_vectors2], [coord[1] for coord in dino_vectors2], 'bo-')
plt.show()
```

```python
# 设置坐标轴范围
plt.xlim(-4, 4)
plt.ylim(-4, 4)

# 绘制坐标轴的水平线和垂直线
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)

# 定义向量 k 和 start_point
k = (1, 3)
start_point = (0, 0)

# 计算向量 dino_vectors3
dino_vectors3 = np.add((-1.5, -2.5), k)

# 绘制向量 k
plt.quiver(start_point[0], start_point[1], k[0], k[1], angles='xy', scale_units='xy', scale=1, color='b')

# 绘制向量 dino_vectors3，起点是向量 k 的终点
plt.quiver(k[0], k[1], dino_vectors3[0], dino_vectors3[1], angles='xy', scale_units='xy', scale=1, color='g')

# 计算向量 k 和 dino_vectors3 的和，并绘制
total_vector = np.add(k, dino_vectors3)
plt.quiver(start_point[0], start_point[1], total_vector[0], total_vector[1], angles='xy', scale_units='xy', scale=1,
		   color='y')

# 显示图形
plt.show()
```

```python
plt.axis((-10, 10, -10, 10))
plt.axhline(0, color='black', linewidth=0.5)
plt.axvline(0, color='black', linewidth=0.5)
k = np.array([1, 3])
startpoint = np.array([0, 0])
p = np.array([1, -3])
# 绘制向量
plt.annotate('', xy=startpoint + k, xytext=startpoint,
			 arrowprops=dict(facecolor='red', arrowstyle="->"))
plt.annotate('', xy=k + p, xytext=k,
			 arrowprops=dict(facecolor='red', arrowstyle="->"))
plt.annotate('', xy=start_point + k + p, xytext=start_point,
			 arrowprops=dict(facecolor='red', arrowstyle="->"))

# 显示图形
plt.show()
```
