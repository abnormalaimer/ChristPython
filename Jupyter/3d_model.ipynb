```python
import numpy as np
import matplotlib.pyplot as plt

cube_vertices = np.array([
	[-1, -1, -1],  # 0
	[-1, -1, 1],  # 1
	[-1, 1, -1],  # 2
	[-1, 1, 1],  # 3
	[1, -1, -1],  # 4
	[1, -1, 1],  # 5
	[1, 1, -1],  # 6
	[1, 1, 1]  # 7
])

# 12
cube_edges = [
	(0, 1), (0, 2), (0, 4),
	(1, 3), (1, 5),
	(2, 3), (2, 6),
	(3, 7),
	(4, 5), (4, 6),
	(5, 7),
	(6, 7)
]

# 3D
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 
for edge in cube_edges:
	# 
	edge_vertices = cube_vertices[list(edge)]  #  list  edge 
	print("Shape of edge_vertices:", edge_vertices.shape)
	print("Content of edge_vertices:", edge_vertices)
	
	#  x, y, z 
	x = edge_vertices[:, 0]
	y = edge_vertices[:, 1]
	z = edge_vertices[:, 2]
	
	# 
	ax.plot(x, y, z, color='b')

# 
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# 
plt.show()
```
