```python
import numpy as np
import matplotlib.pyplot as plt

cube_vertices = np.array([
	[-1, -1, -1],  # 0
	[-1, -1, 1],  # 1
	[-1, 1, -1],  # 2
	[-1, 1, 1],  # 3
	[1, -1, -1],  # 4
	[1, -1, 1],  # 5
	[1, 1, -1],  # 6
	[1, 1, 1]  # 7
])

# 定义立方体的12条边
cube_edges = [
	(0, 1), (0, 2), (0, 4),
	(1, 3), (1, 5),
	(2, 3), (2, 6),
	(3, 7),
	(4, 5), (4, 6),
	(5, 7),
	(6, 7)
]

# 创建图形和3D坐标轴
fig = plt.figure()
ax = fig.add_subplot(111, projection='3d')

# 遍历每条边并绘制
for edge in cube_edges:
	# 提取边的两个顶点
	edge_vertices = cube_vertices[list(edge)]  # 使用 list 函数确保 edge 是一个列表
	print("Shape of edge_vertices:", edge_vertices.shape)
	print("Content of edge_vertices:", edge_vertices)
	
	# 提取 x, y, z 坐标
	x = edge_vertices[:, 0]
	y = edge_vertices[:, 1]
	z = edge_vertices[:, 2]
	
	# 绘制线段
	ax.plot(x, y, z, color='b')

# 设置坐标轴标签
ax.set_xlabel('X axis')
ax.set_ylabel('Y axis')
ax.set_zlabel('Z axis')

# 显示图形
plt.show()
```
